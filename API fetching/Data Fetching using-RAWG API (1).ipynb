```python
import requests
import pandas as pd


# API key from RAWG
api_key = "#############################"

def fetch_game_data(page=1):
    url = f"https://api.rawg.io/api/games?key={api_key}&page={page}"
    response = requests.get(url)
    data = response.json()
    return data

def fetch_game_details(game_id):
    url = f"https://api.rawg.io/api/games/{game_id}?key={api_key}"
    response = requests.get(url)
    data = response.json()
    return data.get("description", "No description available")



game_list = []

# Fetch 250 pages to gather around 5,000 games (20 games per page)
for page in range(1, 251):  # Adjust the range to get more pages
    print(f"Fetching page {page}...")
    data = fetch_game_data(page)

    # Check if the response is valid
    if "results" not in data:
        print(f"Error on page {page}: {data.get('detail', 'Unknown error')}")
        break

    for game in data['results']:
        # Get the detailed description using the game ID
        detailed_description = fetch_game_details(game.get("id"))
        metacritic_url = fetch_game_details(game.get("id"))

        game_info = {
            "game_id": game.get("id"),
            "name": game.get("name"),
            "description": detailed_description,
            "genres": ", ".join([genre['name'] for genre in game.get("genres", [])]),
            "platforms": ", ".join([platform['platform']['name'] for platform in game.get("platforms", [])]),
            "rating": game.get("rating"),
            "released": game.get("released"),
            "cover_image": game.get("background_image"),
            "game_link": f"https://rawg.io/games/{game.get('slug')}",
        }
        game_list.append(game_info)



# Convert to DataFrame and save as CSV
games_df = pd.DataFrame(game_list)
games_df.to_csv("games_metadata_5k.csv", index=False, encoding="utf-8", sep=",")
print("Data saved to games_metadata_5k.csv")
```
